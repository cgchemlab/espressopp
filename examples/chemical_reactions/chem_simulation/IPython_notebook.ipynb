```python
import networkx as nx
```

```python
h5['/connectivity'].keys()
```

```python
cl = h5['/connectivity/chem_fpl_0/value']
```

```python
normal_bonds = [(x+p*6, (x+1)+p*6) for x in range(1, 6) for p in range(0, 120)]
g = nx.Graph()
for b1, b2 in normal_bonds:
    g.add_edge(b1, b2)
```

```python
number_of_components = []
size_of_large_component = []
average_degree = []
for c in cl:
    g1 = g.copy()
    for b1, b2 in c:
        if b1 != -1 and b2 != -1:
            g1.add_edge(b1, b2)
    number_of_components.append(len(list(nx.connected_components(g1))))
    size_of_large_component.append(max([len(x) for x in nx.connected_components(g1)]))
    average_degree.append(np.average(nx.degree(g1).values()))
```
